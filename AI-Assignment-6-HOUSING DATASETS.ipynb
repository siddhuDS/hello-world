```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('D:\\Download_from_Chrome\\house-prices-advanced-regression-techniques\\train.csv')
test = pd.read_csv('D:\\Download_from_Chrome\\house-prices-advanced-regression-techniques\\test.csv')
```

```python
def clean(data):
    data = data.drop(['Alley','FireplaceQu','PoolQC','Fence','MiscFeature'],1)
    data = data. dropna()
    return data
```

```python
train = clean(train)
test = clean(test)
```

```python
Y_train = train['SalePrice']
train = train.drop(['SalePrice'],1)
```

```python
Y_train.head()
```

```python
data = pd.concat([train,test])
```

```python
data.head()
```

```python
categorical = data.select_dtypes(exclude = np.number)
numerical = data.select_dtypes(include = np.number)
categorical= pd.get_dummies(categorical)
X = pd.concat([numerical,categorical],1)
```

```python
X_train = X.iloc[0:1094,:]
X_test = X.iloc[1094:2203,:]
```

```python

from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()

lr_model= lin_reg.fit(X_train,Y_train)
y_pred_lr= lr_model.predict(X_test)
```

```python
y_pred_lr
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state = 42)

dt_model= dt.fit(X_train,Y_train)
y_pred_dt= dt_model.predict(X_test)
```

```python
y_pred_dt
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 42)

rf_model= rf.fit(X_train,Y_train)
y_pred_rf= rf_model.predict(X_test)
```
