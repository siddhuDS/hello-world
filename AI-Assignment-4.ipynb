```python
def clean(col_name):
    m = data[(data[col_name]!='?')][col_name].astype(float).mean()
    data[col_name] = data[col_name].str.replace('?',str(m)).astype(float)
    return data
```

```python
clean('normalized-losses')
clean('bore')
clean('stroke')
clean('horsepower')
clean('peak-rpm')
clean('price')
```

```python
def clean_categorical(column):
    mode = data[data[column]!='?'][column].astype(str).mode()
    data[column] = data[column].str.replace("?",str(mode)).astype(str)
    return data
```

```python
clean_categorical('num-of-doors')
```
