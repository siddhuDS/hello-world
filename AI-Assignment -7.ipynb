```python
df['SEX']=df['SEX'].astype(object)
df['MARRIAGE']=df['MARRIAGE'].astype(object)
df['EDUCATION']=df['EDUCATION'].astype(object)
```

```python
df=pd.get_dummies(df)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
col_names=list(df)
X=scaler.fit_transform(df)
X=pd.DataFrame(X)
X.columns=col_names
```

```python
X=X.drop(['ID'],1)
```

```python
X.head()
```

```python
y=df['default.payment.next.month']
X=X.drop(['default.payment.next.month'],1)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X,y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, recall_score,accuracy_score
from datetime import datetime
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
logscore=logreg.score(X_test,y_test)
print(logscore)
```

```python
crossval_log = cross_val_score(logreg,X_train,y_train,cv=10).mean()
print(crossval_log)
```

```python

ypred_log= logreg.predict(X_test)
log_matrix = confusion_matrix(y_test, ypred_log)
print(log_matrix)
pre_log = precision_score(y_test, ypred_log)
recall_log = recall_score(y_test, ypred_log)
time_log = str(datetime.now())
print(pre_log)
print(recall_log)
```

```python
from sklearn.svm import SVC
svc  = SVC()
svc.fit(X_train,y_train)
ypred_svc= svc.predict(X_test)
accuracy_score(y_test,ypred_svc)
```

```python
crossval_svc = cross_val_score(svc,X_train,y_train,cv=10).mean()
print(crossval_log)
```

```python
svc_matrix = confusion_matrix(y_test, ypred_svc)
print(svc_matrix)
pre_svc = precision_score(y_test, ypred_svc)
recall_svc = recall_score(y_test, ypred_svc)
time_svc = str(datetime.now())
print(pre_svc)
print(recall_svc)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn  = KNeighborsClassifier()
knn.fit(X_train,y_train)
ypred_knn = knn.predict(X_test)
accuracy_score(y_test,ypred_knn)
```

```python
crossval_knn = cross_val_score(knn,X_train,y_train,cv=10).mean()
print(crossval_knn)
```

```python
knn_matrix = confusion_matrix(y_test, ypred_knn)
print(knn_matrix)
pre_knn = precision_score(y_test, ypred_knn)
recall_knn = recall_score(y_test, ypred_knn)
time_knn = str(datetime.now())
print(pre_knn)
print(recall_knn)
```

```python
Model = ["Logistic Regression","Support Vector Classifier", "K-NN Classifier"]
cross_val= [crossval_log,crossval_svc,crossval_knn]
matrix = [log_matrix,svc_matrix,knn_matrix]
precision = [pre_log,pre_svc,pre_knn]
recall = [recall_log,recall_svc,recall_knn]
time = [time_log,time_svc,time_knn]
results=pd.DataFrame({'TimeStamp':time,'Model': Model,'CrossValidation Score': cross_val,'Accuracy Matrix': matrix,'Precision': precision,'Recall': recall})
results
```
